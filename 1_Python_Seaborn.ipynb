# Changing Figure Aesthetic<br>
set_style() method is used to set the aesthetic of the plot. It means it affects things like the color of the axes, whether the grid is active or not, or other aesthetic elements. There are five themes available in Seaborn.

<br>

darkgrid<br>
whitegrid<br>
dark<br>
white<br>
ticks<br>
Syntax:<br>

set_style(style=None, rc=None)

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)

sns.set_style("dark")
plt.show()
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
sns.set_style("darkgrid")
plt.show()
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
sns.set_style("whitegrid")
plt.show()
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
sns.set_style("ticks")
plt.show()
```

# Scaling the plots: <br>
It can be done using the set_context() method. It allows us to override default parameters. This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. The base context is “notebook”, and the other contexts are “paper”, “talk”, and “poster”. font_scale sets the font size.
<br>
Syntax:<br>

set_context(context=None, font_scale=1, rc=None)

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
sns.set_context(context="paper",font_scale=1)
plt.show()
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
sns.set_context(context="talk",font_scale=1)
plt.show()
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=data)
sns.set_context(context="poster",font_scale=1)
plt.show()
```

# **subplot():** <br>
function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function. Another drawback of the subplot function is that it deletes the preexisting plot on your figure.
It is a wrapper of Figure.add_subplot.

```python
#Example-1
x1=[0,1,2,3,4]
y1=[50,20,30,10,40]

x2=[10,20,30,40,50]
y2=[1,2,3,4,5]

plt.figure(figsize=(10,6))
plt.suptitle("sub Plots")

plt.subplot(1,2,1)
plt.title("plot1")
plt.plot(x1,y1)

plt.subplot(1,2,2)
plt.title("plot2")
plt.plot(x1,y1)

plt.show()
```

```python
#Example-2
x3=[10,20,30,40,50]
y3=[0,1,5,6,7]

#figure size
#figsize(width,height)
plt.figure(figsize=(20,10))

plt.suptitle("Sup Figure")

plt.subplot(2,2,1)
plt.title("Plot-1")
plt.plot(x3,y3)

plt.subplot(2,2,2)
plt.title("Plot-2")
plt.plot(x3,y3)

plt.subplot(2,2,3)
plt.title("Plot-3")
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.title("Plot-3")
plt.plot(x3,y3)

plt.show()
```

# **Setting the Style Temporarily :**<br>

axes_style() method is used to set the style temporarily. It is used along with the with statement.
<br>
Syntax:<br>
axes_style(style=None, rc=None)

```python
import seaborn as sns
import matplotlib.pyplot as plt

data=sns.load_dataset("iris")

#setting the figure size
plt.figure(figsize=(8,18))

#setting the title
plt.suptitle("Figure one")

def fun2():
  return sns.lineplot(x="sepal_length",y="sepal_width",data=data)

#inserting subplots

#now setting the axes_style
with sns.axes_style("darkgrid"):
  plt.title("\nplt-1")
  plt.subplot(2,1,1)
  sns.despine()
  fun2()

#now we plot another subplot
plt.subplot(2,1,2)
x=[1,2,3]
y=[10,50,60]
plt.plot(x,y)
sns.despine()
plt.title("Plot-2")

plt.show()
```

# **Multiple plot using seaborn**
Method 1: Using FacetGrid() method

FacetGrid class helps in visualizing distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels.

A FacetGrid can be drawn with up to three dimensions ? row, col, and hue. The first two have obvious correspondence with the resulting array of axes; think of the hue variable as a third dimension along a depth axis, where different levels are plotted with different colors.

FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis

```python
data['species'].unique()
```

```python
data2=sns.load_dataset("tips")
data2
```

```python
data2.head()
```

```python
# Now we use FacetGrid()
import seaborn as sns
import matplotlib.pyplot as plt

var=sns.FacetGrid(data2,height=6,aspect=1,col="sex",hue="day")  # palette="A"---->For diiferent color
var.map(plt.scatter,"total_bill","tip").add_legend()

plt.show()
```

# **Barplot**
A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column

```python
fig=plt.subplots(figsize=(10,10))
sns.barplot(x='sex',y='total_bill',data=data2,palette='plasma')
sns.despine()
sns.set_style('darkgrid')
plt.show()
```

```python
import numpy as np

fig=plt.figure(figsize=(10,8))
sns.barplot(x='sex',y='total_bill',data=data2,palette='plasma',estimator=np.std)
plt.show()
```

# **Countplot**
A countplot basically counts the categories and returns a count of their occurrences. It is one of the most simple plots provided by the seaborn library.

Explanation/Analysis

Looking at the plot we can say that the number of males is more than the number of females in the dataset. As it only returns the count based off a categorical column, we need to specify only the x parameter

```python
data2['sex'].value_counts()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='sex',data=data2,palette='plasma')
plt.show()
```
